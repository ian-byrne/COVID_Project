```python
import pandas as pd
```

```python
df = pd.read_csv('covid_case_data.csv', low_memory=False)
```

```python
#df.head()
```

```python
correct = pd.notnull(df['Country_Region'])
correct = df[correct].copy()
correct.drop(columns=['Province/State', 'Country/Region', 
                      'Last Update', 'Latitude', 'Longitude'
                     ], inplace=True)
```

```python
wrong = pd.isna(df['Country_Region'])

wrong = df[wrong].copy()
wrong.drop(columns=['Combined_Key', 'Province_State', 
                    'Country_Region', 'Lat', 'Long_'], 
           inplace=True
          )
```

```python
col_names = {'Province/State': 'Province_State', 
             'Country/Region': 'Country_Region',
             'Last Update': 'Last_Update',
             'Latitude': 'Lat',
             'Longitude': 'Long_'}
```

```python
wrong.rename(columns=col_names, inplace=True)
```

```python
wrong['Last_Update'] = pd.to_datetime(wrong['Last_Update'])
wrong['Date'] = wrong['Last_Update'].dt.date
wrong.drop(columns='Last_Update', inplace=True)
```

```python
correct.head()
```

```python
new_df = pd.concat([correct, wrong], ignore_index=True, sort=True)
```
